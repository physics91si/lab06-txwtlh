```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
```

# Part 1

```python
# TODO define gaussian function here
def gaussian(x, mean, std):
    return (np.exp(-(x-mean)**2/(2*std**2)))
```

```python
# TODO plot gaussians here
x = np.arange(-5, 5, 0.01)
plt.plot(x, gaussian(x, 2, 0.1), label="mean 2, std 0.1")
plt.plot(x, gaussian(x, 0, 0.4), label="mean 0, std 0.4")
plt.plot(x, gaussian(x, -1, 1), label="mean -1, std 1")
plt.legend()
```

```python
def curve(x, l,m,n):
    return l**(-(x-m)**2/(2*n**2))

# TODO fit gaussians, with known parameters, here
y = gaussian(x, -1, 1)
params, cov = optimize.curve_fit(curve, x, y)

# TODO plot result
plt.plot(x, y, "b-")
plt.plot(x, curve(x, *params), "r--")
```

```python
# TODO: load data here
# dt = np.dtype([('x', float), ('y', float)])
xdata, ydata = np.loadtxt('HDalpha_data', unpack=True)
```

```python
# TODO: plot it
plt.plot(xdata, ydata)
plt.suptitle('Absorption Signal', fontsize=20)
```

```python
# TODO: plot data as emission signal
ydata *= -1;
plt.plot(xdata, ydata)
plt.suptitle('Emission Signal', fontsize=20)
```

```python
# TODO: fit a function to this data and plot it
def emission_function(xvar, a,b,c,d,c1,c2):
    return c1*curve(xvar, np.exp(1),a,b) + c2*curve(xvar, np.exp(1),c,d)

guess = [13128, 1, 13132, 1, 0.5, 0.5]
popt, pcov = optimize.curve_fit(emission_function, xdata, ydata, p0=guess)

# TODO plot result
plt.plot(xdata, ydata, "b-", label="Emission signal")
plt.plot(xdata, emission_function(xdata, *popt), "r-", label="Fitted curve")
plt.legend()
plt.suptitle("Emission Signal of Hydrogen/Deuterium", fontsize=20)
```

# Part 3

```python
def flux_for_dipole(x_pos, y_pos, Q_vals, X, Y):
    """
    Returns the flux based on the information of the charges
    and the grid of the plot.
    Code modified from Physics 113, written by Blas Cabrera
    """
    flux = np.zeros((len(X), len(Y)))
    x_m = np.outer(np.ones((X.size)), X); 
    y_m = np.outer(Y, np.ones((Y.size))); 
    for n in range(len(Q_vals)):
        flux += 2*np.pi*Q_vals[n]*(1 - (x_pos[n]-x_m)/np.sqrt((y_pos[n]-y_m)**2+(x_pos[n]-x_m)**2))
    return flux
```

```python
# TODO create arrays of information
px = [-5, 5]
py = [5, 2]
q = [1.602e-19, 2.12e-19]
arr_X = np.arange(-31.5, 32)
arr_Y = np.arange(-31.5, 32)

# TODO get flux
flux = flux_for_dipole(px, py, q, arr_X, arr_Y)

# TODO plot contours of the flux
area = [200, 200]
plt.scatter(px, py, s=area, c='black')
plt.contour(arr_X, arr_Y, flux, 20)
plt.suptitle('Electric Field', fontsize = 20)
plt.xticks([])
plt.yticks([])
```
